## **1. dataset:  titanic.csv**

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
# membaca dataset 
dataset = pd.read_csv("titanic.csv")
dataset
```

# Merubah fitur sex yang awalnya teks menjadi numerik

```python
le_sex = LabelEncoder()
dataset['Sex'] = le_sex.fit_transform(dataset['Sex'])
dataset['Sex']
```

**ANALISIS**
melakukan encoding menggunakan labelencoder dari library sklearn untuk merubah data yang awalnya teks menjadi numerik, dimana male menjadi angka 1 dan female menjadi angka 0

## **2. data: mengaambil dataset kolom fitur (Sex, Age, Pclass, Fare, Survived)**

```python
# membuat variabel data. Variabel data berisikan beberapa fitur saja dari dataset utama
data = dataset[['Sex', 'Age', 'Pclass', 'Fare', 'Survived']].copy()
data
```

## **3.  train_data: mengambil fitur (Sex, Pclass, Fare, Survived) pada data yang Age tidak sama dengan null**

```python
# pada variabel train_data ini berisikan fitur age yang tidak ada data null
train = data.dropna()
train = train.reset_index(drop=True)
train_data = train.drop(columns=['Age'])
train_data
```

**ANALISIS**

Pertama disini saya akan menghapus data yang missing dengan dropna setelah itu mengatur ulang indeks setelah penghapusan,jadi indeks data yang telah dihapus tidak akan di masukkan di kolom baru,sehingga urutanya numeiknya urut tanpa mempertahankan indeks data yang telah di hapus 

Karena di data_train hanya menggunakan fitur pclass,fare,survived maka fitur age dihapus

## **4. train_label: mengambil fitur (Age) pada data yang Age tidak sama dengan null**

```python
# variabel train_label ini mengandung nilai yang tidak null pada fitur age
train_label = train['Age']
train_label
```

**ANALISIS**

variabel train sebelumnya adalah variabel yang menyimpan data yang sudah tidak mengandung nilai null, di sini saya membuat variabel train_label yang hanya berisikan fitur age saja yang tidak ada nilai null nya

## **6. train_data: lakukan normalisasi pada train_data dengan Min-Max 0-1 (catat nilai min dan max setiap atribut)**

```python
from sklearn.preprocessing import MinMaxScaler
minmax = MinMaxScaler()
train_data_norm = minmax.fit_transform(train_data)
train_data_norm
```

**ANALISIS**

tahap  ini akan melakukan standarisasi dengan minmax setelah itu akan menghitung nilai min dan maks setiap fitur pada data train_data dengan fungsi fit_transform() dimana fit untuk  menghitung nilai minimum dan maksimum dan transform() untuk normalisasi menggunakan nilai min dan maks yang telah didapatkan

# Menampilka nilai min dan max setiap atribut setelah dilakukan normalisasi

```python
min_sex = train_data_norm[:, 0].min()
print(f'Nilai minimum atribut Sex setelah normalisasi: {min_sex}')
max_sex = train_data_norm[:, 0].max()
print(f'Nilai maximum atribut Sex setelah normalisasi: {max_sex}')

min_age = train_data_norm[:, 1].min()
print(f'Nilai minimum atribut Age setelah normalisasi: {min_age}')
max_age = train_data_norm[:, 1].max()
print(f'Nilai maximum atribut Age setelah normalisasi: {max_age}')

min_pclass = train_data_norm[:, 2].min()
print(f'Nilai minimum atribut Pclass setelah normalisasi: {min_pclass}')
max_pclass = train_data_norm[:, 2].max()
print(f'Nilai maximum atribut Pclass setelah normalisasi: {max_pclass}')

min_fare = train_data_norm[:, 3].min()
print(f'Nilai minimum atribut Fare setelah normalisasi: {min_fare}')
max_fare = train_data_norm[:, 3].max()
print(f'Nilai maximum atribut Fare setelah normalisasi: {max_fare}')
```

## **7.  test_data: lakukan normalisasi pada test_data dengan Min-Max 0-1 (dengan nilai min dan max setiap atribut pada Langkah 6)**

```python
test_data_norm = minmax.transform(test_data)
test_data_norm
```

## **8. class_result: Lakukan klasifikasi test_data terhadap train_data dengan 3-NN (output mepakai class pada train_label)**

```python
from sklearn.neighbors import KNeighborsRegressor
knn_model = KNeighborsRegressor(n_neighbors=3)
knn_model.fit(train_data_norm, train_label)
class_result = knn_model.predict(test_data_norm)
class_result
```

**ANALISIS**

Pada tahap ini dilakukan klasifikasi dengan metode knn dengan menggunakan n sebanyak 3 artinya Model ini menggunakan tiga tetangga terdekat untuk memprediksi nilai output untuk data uji.

## **11. train_data: ambil fitur (Sex, Age, Pclass, Fare) dari data**

```python
train_data = data[['Sex', 'Age', 'Pclass', 'Fare']]
train_data
```

## **12.  train_label: ambil fitur (Survived) dari data**

```python
train_label = data['Survived']
train_label
```

## **13. test_data: ambil test_dataset kolom fitur (Sex, Age, Pclass, Fare). Hilangkan baris data yang terdapat missing values**

```python
test_data = test_dataset[['Sex', 'Age', 'Pclass', 'Fare']]
test_data = test_data.dropna()
test_data
```

## **14. test_label: titanic_testlabel.csv (urutan sesuai test_data no.13)**

```python
test_label = pd.read_csv("titanic_testlabel.csv")['Survived']
test_label = test_label.loc[test_data.index]
test_label
```

## **15. train_data: lakukan normalisasi pada train_data dengan Min-Max 0-1 (catat nilai min dan max setiap atribut)**

```python
minmax2 = MinMaxScaler()
train_data_norm = minmax2.fit_transform(train_data)
train_data_norm
```

```python
min_sex = train_data_norm[:, 0].min()
print(f'Nilai min atribut Sex setelah normalisasi: {min_sex}')
max_sex = train_data_norm[:, 0].max()
print(f'Nilai max atribut Sex setelah normalisasi: {max_sex}')

min_age = train_data_norm[:, 1].min()
print(f'Nilai min atribut Age setelah normalisasi: {min_age}')
max_age = train_data_norm[:, 1].max()
print(f'Nilai max atribut Age setelah normalisasi: {max_age}')

min_pclass = train_data_norm[:, 2].min()
print(f'Nilai min atribut Pclass setelah normalisasi: {min_pclass}')
max_pclass = train_data_norm[:, 2].max()
print(f'Nilai max atribut Pclass setelah normalisasi: {max_pclass}')

min_fare = train_data_norm[:, 3].min()
print(f'Nilai min atribut Fare setelah normalisasi: {min_fare}')
max_fare = train_data_norm[:, 3].max()
print(f'Nilai max atribut Fare setelah normalisasi: {max_fare}')
```

## **16. test_data: lakukan normalisasi pada test_data dengan Min-Max 0-1 (dengan nilai min dan max setiap atribut pada Langkah 15)**

```python
test_data_norm = minmax2.transform(test_data)
test_data_norm
```

## **17. class_result: Lakukan klasifikasi test_data terhadap train_data dengan 3-NN (output mepakai class pada train_label)**

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=3)
knn_model.fit(train_data_norm, train_label)
class_result = knn_model.predict(test_data_norm)
class_result
```

## **18. error: Bandingkan hasil klasifikasi class_result dengan test_label. Jika tidak sama berarti error. Hitunglah jumlah error dari seluruh class_result**

```python
error = (class_result != test_label.values.ravel()).sum()
print( 'didapatkan error sebanyak:', error)
```

## **19. error_ratio: error dibagi jumlah test_data, dikali 100 (%)**

```python
error_ratio = (error / len(test_data)) * 100
print("Error ratio: {:.2f}%".format(error_ratio))
```
