# **2. Assignment: Portugal**
## **2.1. Dataset**

```python
import pandas as pd

dataset = pd.read_csv('transaction.csv')
dataset.head()
```

```python
print(dataset.isna().sum())
```

## **2.2. Data: Portugal**

```python
data = dataset.loc[(dataset['Country'] == 'Portugal')].reset_index(drop=True)
data.head()
```

**ANALISIS**

Disini saya menampilkan data transaksi pada negara portugal saja

## **2.3. Transaksi: Portugal**

```python
transaksi = data.groupby('InvoiceNo')['StockCode'].apply(list).reset_index()
transaksi
```

**ANALISIS**

Setelah itu saya mengelompokkan data berdasarkan InvoiceNo dan menggabungkan daftar StockCode yng ada disetiap transaksi menjadi 1 baris

Pertama di sini saya akan  mengelompokkan data berdasarkan InvoiceNo dengan menggunakan groupby kemudian Dari setiap grup, kita ambil kolom StockCode selanjutnya merubah mnejadi list agar semua produk dalam satu transaksi jadi satu daftar.

## **2.4. Association Rule: Portugal**

```python
from mlxtend.preprocessing import TransactionEncoder
# merubah format menjadi biner
te = TransactionEncoder() 
te_ary = te.fit(transaksi['StockCode']).transform(transaksi['StockCode'])
transaksi = pd.DataFrame(te_ary, columns=te.columns_)

# menjalankan algoritma apriori
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

frequent_itemsets=apriori(transaksi, min_support=0.2, use_colnames=True)

#Menghasilkan Aturan Asosiasi
rules=association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)

rules[['antecedents', 'consequents', 'confidence']]
```

**ANALISIS**

Disini saya melakukan assosiation rule untuk langkah-langkahnya yakni :
1. Mengubah data transaksi dari format list produk per transaksi menjadi format biner (0/1) agar bisa diproses oleh algoritma apriori.

2.  selanjutnya menjalankan lagoritma apriori, di sini min_support di atur mnejadi 0.2 yang artinya mengambil data yang muncul pada minimal 20% transaksi.

3. Menghasilkan Aturan Asosiasi = pada aturan asosiasi ini saya memberikan confidence minimal 70% artinya Hanya tampilkan aturan-aturan yang punya tingkat kepercayaan minimal 70%.Aturan seperti "jika beli A → beli B" akan ditampilkan kalau confidence-nya ≥ 70%.


# **3. Assignment: Greece**
## **3.1. Dataset**

```python
import pandas as pd

dataset = pd.read_csv('transaction.csv')
dataset
```

## **3.2. Data: Greece**

```python
data = dataset.loc[(dataset['Country'] == 'Greece')].reset_index(drop=True)
data.head()
```

## **3.3. Transaksi: Greece**

```python
transaksi = data.groupby('InvoiceNo')['StockCode'].apply(list).reset_index()
transaksi
```

**ANALISIS**

Setelah itu saya mengelompokkan data berdasarkan InvoiceNo dan menggabungkan daftar StockCode yng ada disetiap transaksi menjadi 1 baris

Pertama di sini saya akan  mengelompokkan data berdasarkan InvoiceNo dengan menggunakan groupby kemudian Dari setiap grup, kita ambil kolom StockCode selanjutnya merubah mnejadi list agar semua produk dalam satu transaksi jadi satu daftar.

## **3.4. Association Rule: Greece**

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(transaksi['StockCode']).transform(transaksi['StockCode'])
transaksi = pd.DataFrame(te_ary, columns=te.columns_)

from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

frequent_itemsets=apriori(transaksi, min_support=0.25, use_colnames=True)
rules=association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)

rules[['antecedents', 'consequents', 'confidence']]
```

**ANALISIS**

Disini saya melakukan assosiation rule untuk langkah-langkahnya yakni :
1. Mengubah data transaksi dari format list produk per transaksi menjadi format biner (0/1) agar bisa diproses oleh algoritma apriori.

2.  selanjutnya menjalankan lagoritma apriori, di sini min_support di atur menjadi 0.25 yang artinya mengambil data yang  muncul dalam minimal 25% dari seluruh transaksi 

3. Menghasilkan Aturan Asosiasi = pada aturan asosiasi ini saya memberikan confidence minimal 50% artinya Hanya tampilkan aturan-aturan yang punya tingkat kepercayaan minimal 50%.Aturan seperti "jika beli A → beli B" akan ditampilkan kalau confidence-nya ≥ 50%.
